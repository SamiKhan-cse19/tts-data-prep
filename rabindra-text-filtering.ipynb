```python
import glob
import os
from bnlp import NLTKTokenizer

bnltk = NLTKTokenizer()
```

```python
rabindra_texts = []
text_files = glob.glob('Rabindra_Rachnabali/*/*.txt', recursive=True)
for file in text_files:
  with open(file, 'r') as f:
    text = f.read()
    for paragraph in text.split('\n'):
      paragraph = paragraph.strip()
      sentence_tokens = bnltk.sentence_tokenize(paragraph)
      for sentence in sentence_tokens:
        if len(sentence) > 100:
          rabindra_texts.append(sentence.strip())

len(rabindra_texts)
```

```python
with open('rabindra_corpus.txt', 'w') as f:
  for text in rabindra_texts:
    f.write(text + '\n')
```

```python
# length distribution of sentences
lengths = [len(sentence) for sentence in rabindra_texts]
import matplotlib.pyplot as plt
plt.hist(lengths, bins=50)
plt.show()
```
